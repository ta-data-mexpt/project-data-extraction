```python
import requests
import time
import pandas as pd
```

```python
req = requests.get('https://digi-api.com/api/v1/digimon?pageSize=1000')
req.status_code
```

```python
digimon_json = req.json()
digimon_json
```

```python
# primer digimon de la lista
digimon_json['content'][0]
```

```python
digi_list = []
for i in range (0,999):
    digi_list.append(digimon_json['content'][i])  

digi_list
```

```python
# primeros 1000 digimon de la API

digi_df = pd.DataFrame(digi_list)
digi_df
```

```python
digi_df.to_csv('1000_digimon.csv')
```
