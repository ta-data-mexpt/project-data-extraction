```python
import json
import requests
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
from urllib.request import urlopen
```

```python
# Datos Generales pokemon

df=pd.DataFrame()


for i in range(1,151):
    url = f'https://pokeapi.co/api/v2/pokemon/{i}/'
    response = requests.get(url)
    if response.status_code == 200:
           
        results = response.json()
        flattened_data = json_normalize(results)
        df = df.append(flattened_data)
        
    else:
        print('connection error')
        break
print('Connecion Successful')
```

```python
df.index = np.arange(1,len(df)+1)
```

```python
ids = [df['id'][i] for i in range(1,151)]
```

```python
names = [df['forms'][i][0]['name'] for i in range(1,151)]
```

```python
attacks = [df['moves'][i][0]['move']['name'] for i in range(1,151)]
```

```python
links_pictures = [df['sprites.versions.generation-vi.x-y.front_shiny'][i] for i in range(1,151)]
```

```python
# Para obtener el tipo de pokemon

df2=pd.DataFrame()


for i in range(1,151):
    url = f'https://pokeapi.co/api/v2/pokemon-form/{i}/'
    response = requests.get(url)
    if response.status_code == 200:
        
        results = response.json()
        flattened_data = json_normalize(results)
        df2 = df2.append(flattened_data)
    else:
        print('connection error')
        break

print('Connecion Successful')
```

```python
df2.index = np.arange(1,len(df2)+1)
```

```python
types = [df2['types'][i][0]['type']['name'] for i in range(1,151)]
```

```python
# Para obtener las fotos de pokemons por la parte de atr√°s

df3=pd.DataFrame()

for i in range(1,151):
    url = f'https://pokeapi.co/api/v2/pokemon-form/{i}/'
    response = requests.get(url)
    if response.status_code == 200:
           
        results = response.json()
        flattened_data = json_normalize(results)
        flattened_data = flattened_data.set_index('id')
        df3 = df3.append(flattened_data)
        
    else:
        print('connection error')
        break
print('Connecion Successful')
```

```python
df3.index = np.arange(1,len(df3)+1)
```

```python
links_pictures_back = [df3['sprites.back_default'][i] for i in range(1,151)]
```

```python
nom_columns = ['Id','Pokemon','Type','Hard Attack','Picture','Back _Picture']
```

```python
df_pokemon = pd.DataFrame(list(zip(ids, names, types,  attacks, links_pictures, links_pictures_back )), columns = nom_columns)
```

```python
df_pokemon = df_pokemon.set_index('Id')
```
