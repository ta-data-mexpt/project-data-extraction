DataFrame
-------

```python
# un solo df
df_todo = pd.concat([df_1, df_1BC,
                     df_2, df_2BC,
                     df_3, df_3BC,
                     df_4, df_4BC,
                     df_5, df_5BC,
                     df_6, df_6BC,
                     df_7, df_7BC,
                     df_8, df_8BC])

df_todo
```

```python
# remove 'enlace int'
col_enlace = df_todo.pop('enlace_int')
```

```python
# agregar 'enlace_int' al principio
df_todo.insert(0, 'enlace_int', col_enlace)
```

```python
df_todo.tail()
```

```python
# df solo de las columnas de nombre del enlace internacional ('enlace_int')
# fecha, hora y
# totales de importación ('imp_total') y exportación ('exp_total')

df_imp_exp = df_todo[['enlace_int', 'fecha', 'hora', 'imp_total', 'exp_total']]
df_imp_exp.head()
```

```python
df_imp_exp.tail()
```

```python
# cambio de Dtype de 'imp_total' y 'exp_total' a int
df_imp_exp['imp_total']  = df_imp_exp['imp_total'].astype(dtype='int64')
```

```python
df_imp_exp['exp_total'] = df_imp_exp['exp_total'].astype(dtype='int64')
```

```python
df_imp_exp.info()
```

```python
# rename columnas

df_julio = df_imp_exp.rename(columns = {
                            'enlace_int': 'Enlace Internacional',
                             'fecha': 'Fecha',
                             'hora': 'Hora',
                             'imp_total': 'Cantidad de Energia Asignada de Importacion (MWh)',
                             'exp_total': 'Cantidad de Energia Asignada de Exportacion (MWh)'});
```

```python
df_julio.head(10)
```

```python
# Archivo csv

df_julio.to_csv('csv/API_julio_imp_exp.csv', index=False)
```

--------

Resumen información
-------

**Suma de todos los enlaces internacionales:**

```python
# suma de todos enlaces
df_suma_enlaces = df_julio.groupby(['Enlace Internacional']).sum()
df_total = df_suma_enlaces.append(df_suma_enlaces.sum().rename("TOTAL"))
df_total
```

**Suma de todos los enlaces ordenados por importaciones:**

```python
# enlaces ordenados por importación mayor a menor
df_imp = df_julio.groupby(['Enlace Internacional']).sum().sort_values(by='Cantidad de Energia Asignada de Importacion (MWh)', ascending=False)
df_suma_imp = df_imp.append(df_imp.sum().rename("TOTAL"))
df_suma_imp
```

**Suma de todos los enlaces ordenados por exportaciones:**

```python
# enlaces ordenados por exportación mayor a menor
df_exp = df_julio.groupby(['Enlace Internacional']).sum().sort_values(by='Cantidad de Energia Asignada de Exportacion (MWh)', ascending=False)
df_suma_exp = df_exp.append(df_exp.sum().rename("TOTAL"))
df_suma_exp
```
