```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Autoriza la API de Google Maps
with open('../../client.txt') as f:
    YOUR_API_KEY = f.readlines()
    YOUR_API_KEY = YOUR_API_KEY[0].split()
    YOUR_API_KEY = YOUR_API_KEY[0]
```

```python
#If your request is malformed or supplies an invalid signature, 
#the Places API returns an HTTP 403 (Forbidden) error.
url = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=-33.8670522,151.1957362&radius=500&types=food&name=harbour&key={YOUR_API_KEY}'
html = requests.get(url)
#Verificar el acceso
print(html)
```

```python
#Probando que funciona
html = requests.get(url).content
soup = BeautifulSoup(html,'html')
print(soup.prettify())
```

```python
output = 'json'
input_text = 'BBVA'
input_type = 'textquery'
field = 'name,place_id,geometry,formatted_address,business_status'
types = 'bank'
```

```python
url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/{output}?input={input_text}&inputtype={input_type}&fields={field}&type={types}&key={YOUR_API_KEY}'
html = requests.get(url).content
soup = BeautifulSoup(html,'html')
print(soup.prettify())
```
