```python
#antes de este paso hay que hacer pip install wptools y wikipedia para poder ejecutar el programa

import json
import wptools
import wikipedia
import pandas as pd
import urllib.request
```

```python
#importando un archivo csv con la lista de companias Fortune 500

url = 'https://raw.githubusercontent.com/MonashDataFluency/python-web-scraping/master/data/fortune_500_companies.csv'

urllib.request.urlretrieve(url, 'fortune_500_companies.csv')
```

```python
ncompany =  'fortune_500_companies.csv'
df= pd.read_csv(ncompany)
```

```python
#especificamos el numero de companias y utilizamos iloc para enumarar y copiar en un data frame nuevo, convirtiendo a lista el data frame resultante

num_of_companies=50
df2=df.iloc[:num_of_companies,:].copy()
companies=df2['company_name'].tolist()
```

```python
#ahora queremos anexar el articulo de wikipedia a cada nombre de la company

wiki_search = [{company : wikipedia.search(company)} for company in companies]

#mientras estaba haciendo el proyecto me di cuenta que es posible que Wikipedia cuente a Apple literalmente como la fruta, entonces para asegurarnos de que sea la empresa:

companies[companies.index('Apple')] = 'Apple Inc.'
```

```python
wiki_data = []
#como queremos hacer match con las infos contenidas en las wiki boxes, hacemos una lista de los elementos que queremos checar

elementos=['founder','location_country','revenue','operating_income','net_income','assets','equity','type','industry','products','num_employees']
```

```python
for company in companies:
    page=wptools.page(company)
    try:
        page.get_parse()#aqui se llama a la API para coloar las datos
        if page.data['infobox']!= None:
            infobox=page.data['infobox']
            data = {elemento : infobox[elemento] if elemento in infobox else '' for elemento in elementos}
        else:
            data = {elemento : '' for elemento in elementos}

        data['company_name'] = company
        wiki_data.append(data)
    except KeyError:
        pass
```

```python
with open('info.json', 'w') as file:
    json.dump(wiki_data,file)
```
