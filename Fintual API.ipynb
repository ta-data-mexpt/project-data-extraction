```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
from datetime import datetime
import re
from bs4 import BeautifulSoup
```

```python
url_web_scraping = "https://fintualist.com/chile/noticias/el-api-de-fintual/"
```

```python
response = requests.get(url_web_scraping)
```

```python
response.status_code
```

```python
results = BeautifulSoup(response.content)
```

```python
results_list = [ texto.text for texto in results.select('div.post-article > p') ]
```

```python
for element in results_list:
    if re.findall(r"\bid: [0-9]", element):
        print(element)
```

```python
fintualfunds = {'conservative_clooney': 188, 'moderate_pit': 187, 'risky_norris': 186, 'very_conservative_streep': 15077}
```

```python
url_base = 'https://fintual.cl/api/real_assets/'
```

```python
startdate = '2019-01-01'
```

```python
enddate = datetime.today().strftime('%Y-%m-%d')
```

## Conservative Cloney

```python
conservative_clooney_id = [ value for key, value in fintualfunds.items() if key == 'conservative_clooney']

df_conservative_clooney = pd.DataFrame()

url_conservative_clooney = f"{url_base}{conservative_clooney_id[0]}/days?from_date={startdate}&to_date={enddate}"
response_conservative_clooney = requests.get(url_conservative_clooney)
results_conservative_clooney = response_conservative_clooney.json()

for info_conservative_clooney in results_conservative_clooney['data']:
    flattened_data_conservative_clooney = pd.json_normalize(info_conservative_clooney)
    flattened_data_conservative_clooney = flattened_data_conservative_clooney.set_index('id')
    df_conservative_clooney = df_conservative_clooney.append(flattened_data_conservative_clooney)
    
df_conservative_clooney.head(3)
```

## Moderate Pit

```python
moderate_pit_id = [ value for key, value in fintualfunds.items() if key == 'moderate_pit']

df_moderate_pit = pd.DataFrame()

url_moderate_pit = f"{url_base}{moderate_pit_id[0]}/days?from_date={startdate}&to_date={enddate}"
response_moderate_pit = requests.get(url_moderate_pit)
results_moderate_pit = response_moderate_pit.json()

for info_moderate_pit in results_moderate_pit['data']:
    flattened_data_moderate_pit = pd.json_normalize(info_moderate_pit)
    flattened_data_moderate_pit = flattened_data_moderate_pit.set_index('id')
    df_moderate_pit = df_moderate_pit.append(flattened_data_moderate_pit)
    
df_moderate_pit.head(3)
```

## Risky Norris

```python
risky_norris_id = [ value for key, value in fintualfunds.items() if key == 'risky_norris']

df_risky_norris = pd.DataFrame()

url_risky_norris = f"{url_base}{risky_norris_id[0]}/days?from_date={startdate}&to_date={enddate}"
response_risky_norris = requests.get(url_risky_norris)
results_risky_norris = response_risky_norris.json()

for info_risky_norris in results_risky_norris['data']:
    flattened_data_risky_norris = pd.json_normalize(info_risky_norris)
    flattened_data_risky_norris = flattened_data_risky_norris.set_index('id')
    df_risky_norris = df_risky_norris.append(flattened_data_risky_norris)
    
df_risky_norris.head(3)
```

## Very Conservative Steep

```python
very_conservative_streep_id = [ value for key, value in fintualfunds.items() if key == 'very_conservative_streep']

df_very_conservative_streep = pd.DataFrame()

url_very_conservative_streep = f"{url_base}{very_conservative_streep_id[0]}/days?from_date={startdate}&to_date={enddate}"
response_very_conservative_streep = requests.get(url_very_conservative_streep)
results_very_conservative_streep = response_very_conservative_streep.json()

for info_very_conservative_streep in results_very_conservative_streep['data']:
    flattened_data_very_conservative_streep = pd.json_normalize(info_very_conservative_streep)
    flattened_data_very_conservative_streep = flattened_data_very_conservative_streep.set_index('id')
    df_very_conservative_streep = df_very_conservative_streep.append(flattened_data_very_conservative_streep)
    
df_very_conservative_streep.head(3)
```

## Summary

```python
df_fintual_funds = pd.concat([df_conservative_clooney, df_moderate_pit, 
                          df_risky_norris, df_very_conservative_streep])
df_fintual_funds.shape
```
