```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import csv
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:R%C3%A9cords_ol%C3%ADmpicos_en_nataci%C3%B3n'
```

```python
response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
print(soup)
```

```python
table = soup.find('table',{'class':'wikitable plainrowheaders'}).tbody
#print(table)

rows = table.find_all('tr')
columns = [v.text.replace('\n','') for v in rows[0].find_all('th')]
#print(columns)
```

```python
df = pd.DataFrame(columns=columns)

for i in range(1, len(rows)):
    tds=rows[i].find_all('td')
    
    if len(tds)==4:
        values = [tds[0].text, tds [1].text,'',tds[2].text, tds[3].text.replace('\n', ''.replace('\xa0',''))]
    else: 
        values = [td. text.replace ('\n','').replace ('\xa0','') for td in tds]
        
    df = df.append(pd.Series(values, index=columns),ignore_index=True)
    #print(df) 
    
    df.to_csv(r'/Users/polly/Documents/Ironhack/Modulo_1/Proyecto_2' + '\\RécordsOlímpicosNatación.csv', index=False)
```
