# Proyecto 2: Web Scraping 

Web Scraping is used to know the WTA (Women's Tennis Association) Classification. This Classification belongs to the woman tennis player world ranking publishes by the WTA since 1975. 

This Classification is updated every week and is the result of the tournaments that have taken place in the last 52 weeks.

**Ranking:** up to August 15th 2022.

```python
import re
import requests

import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
```

Requests
-------

```python
url = 'https://en.wikipedia.org/wiki/WTA_rankings'

response = requests.get(url)
response
```

```python
contenido_wta = response.content
```

Soup
-------

```python
soup = BeautifulSoup(contenido_wta, "html.parser") 
soup
```

```python
ranking_wta = soup.find_all("table", class_="wikitable nowrap sortable")
ranking_wta
```

## 1. Singles

```python
# there are 2 tables with the previous class: singles y doubles individual

# first table: singles
ranking_single = ranking_wta[0]
ranking_single
```

### Ranking

```python
all_rankings = ranking_single.find_all('th')
all_rankings
```

```python
ranking_number = [number.text for number in all_rankings]
```

```python
ranking_number = [ranking_number.strip() for ranking_number in ranking_number]
ranking_number
```

```python
current_ranking = ranking_number[5:]
current_ranking
```

### Player names

```python
all_ranking = ranking_single.find_all('td')
all_ranking
```

```python
ranking_name = [name.text for name in all_ranking]
ranking_name
```

```python
ranking_name = [name_ranking.strip() for name_ranking in ranking_name]
ranking_name
```

```python
names_string = str(ranking_name)
names_string
```

```python
names_clean = names_string.replace('\\xa0', '')
names_clean
```

```python
pattern_names = r'[a-zA-Z].+?[)]'

names_nacionality = re.findall(pattern_names, names_clean)
names_nacionality
```

```python
names_nacionality_string = str(names_nacionality)
```

```python
pattern_each_name = r'\b[A-Z]+\w+\s+\w.+?\b'

names_player = re.findall(pattern_each_name, names_nacionality_string)
names_player
```

### Nationality

```python
pattern_nac = r'[A-Z]+\b'
nacionality = re.findall(pattern_nac, names_nacionality_string)
nacionality
```

### Current scores

```python
pattern_scores = r'\d{1}[,]\d{3}'

current_score = re.findall(pattern_scores, names_clean)
current_score
```

DataFrame
----

```python
df_ranking_single = pd.DataFrame(list(zip(current_ranking, names_player, nacionality, current_score)),
             columns=['Ranking', 'Name Player', 'Nacionality', 'Current Points'])

df_ranking_single
```

```python
# csv

df_ranking_single.to_csv(r'csv/WebScraping_RankingSingle.csv', index=False)
```
