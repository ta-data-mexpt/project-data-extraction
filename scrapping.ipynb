```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains                    
import re
```

```python
pages_to_scrape=4
url_pattern= 'https://www.gandhi.com.mx/libros/los-mas-leidos?p=%s'
def libros_links(content):
    soup=BeautifulSoup(content,"lxml")
    text = [element.get("href") for element in soup.select("a.product-item-link")]
    return text

def scrape_url(url):
    response = requests.get(url)
    result = libros_links(response.content)
    return result
        
links=[]
for i in range(1, pages_to_scrape+1):
    links.append(scrape_url(url_pattern % i))
```

```python
for i in range(len(links)):
    links[i].remove(links[i][-1])
        
```

```python
links1=[link for lista in links for link in lista ]
links1
```

```python

titulo=[]
autor=[]
editorial=[]
precio=[]
edicion=[]
formato=[]
idioma=[]
paginas=[]
año=[]
for link in links1:
    response = requests.get(link)
    result =BeautifulSoup(response.content,"lxml")
    tit= [element.text for element in result.select("h1.page-title>span")]
    titulo.append(tit[0])
    aut=[element.text.strip() for element in result.select("div.autor>a")]
    autor.append(aut[0])
    edit=[element.text.strip() for element in result.select("div.editoriales>a")]
    editorial.append(edit[0])
    pre= [element.text for element in result.select("span.price")]
    if len(pre)==2:
        precio.append(pre[1])
    else:
        precio.append(pre[0])
    
    datoslib= [element.text for element in result.select("span.attr-data")]
    edicion.append(datoslib[0])
    if len(datoslib)==5:
        formato.append(datoslib[1])
        idioma.append(datoslib[2])
        paginas.append(datoslib[3])
        año.append(datoslib[4])
    else: 
        formato.append("null")
        idioma.append(datoslib[1])
        paginas.append(datoslib[2])
        año.append(datoslib[3])
    
```

```python
df=pd.DataFrame()
df["Libreria"]=["Gandhi" for x in range(0,63)]
df["Tipo"]=["Libro" for x in range(0,63)]
df["#Topxtipo"]=[x for x in range(1,64)]
df["Titulo"]=titulo
df["Autor"]=autor
df["Editorial"]=editorial
df["Edicion"]=edicion
df["Formato"]=formato
df["Idioma"]=idioma
df["Paginas"]=paginas
df["Año"]=año
df["Precio"]=precio
```
