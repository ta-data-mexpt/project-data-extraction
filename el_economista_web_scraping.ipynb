```python
import re
import requests
import xml.etree.ElementTree as ET

import pandas as pd

from bs4 import BeautifulSoup
```

```python
url_rss = 'https://www.eleconomista.com.mx/rss/ultimas-noticias'
```

```python
resp = requests.get(url_rss)
```

```python
resp.status_code
```

```python
noticias_economista = resp.content
print(noticias_economista)
```

```python
tree = ET.fromstring(noticias_economista)
tree
```

```python
for element in tree:
    for subelement in element:
        for subsubelement in subelement:
                print(subsubelement.text)
```
