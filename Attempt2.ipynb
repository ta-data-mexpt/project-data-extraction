```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pandas.io.json import json_normalize
import base64
```

```python
url_base = 'https://fintual.cl/api/'
```

## BANKS

```python
url_banks = url_base+'banks'
```

```python
response_banks = requests.get(url_banks)
```

```python
response_banks.status_code
```

```python
results_banks = response_banks.json()
```

```python
df_banks = pd.DataFrame()
for bank in results_banks['data']:
    flattened_data = pd.json_normalize(bank)
    flattened_data = flattened_data.set_index('id')
    df_banks = df_banks.append(flattened_data)
df_banks.head(3)
```

## CONCEPTUAL ASSETS

```python
url_conceptual_assets = url_base+'conceptual_assets/'
```

```python
response_conceptual_assets = requests.get(url_conceptual_assets)
```

```python
response_conceptual_assets.status_code
```

```python
results_conceptual_assets = response_conceptual_assets.json()
```

```python
df_conceptual_assets = pd.DataFrame()
for conseptualassets in results_conceptual_assets['data']:
    flattened_data = pd.json_normalize(conseptualassets)
    flattened_data = flattened_data.set_index('id')
    df_conceptual_assets = df_conceptual_assets.append(flattened_data)
df_conceptual_assets.head(3)
```

```python
conceptual_assets_id_list = list(df_conceptual_assets.index)
len(conceptual_assets_id_list)
```

```python
url_real_assets = []
df_real_assets = pd.DataFrame()

for id_ in conceptual_assets_id_list:
    urlrealassets = url_conceptual_assets+id_+'/real_assets'
    url_real_assets.append(urlrealassets) 
    
for url_ in url_real_assets:
    rspns = requests.get(url_)
    rslts = rspns.json()
        
    for info_rslts in rslts['data']:
        flat_data_real_assets = pd.json_normalize(info_rslts)
        flat_data_real_assets = flat_data_real_assets.set_index('id')
        df_real_assets = df_real_assets.append(flat_data_real_assets)

df_real_assets
```

```python
real_assets_id_list = list(df_real_assets.index)
```
