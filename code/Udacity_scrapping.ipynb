```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url_Udacity = 'https://www.udacity.com/courses/all'
html_Udacity = requests.get(url_Udacity).content
soup_Udacity = BeautifulSoup(html_Udacity)
```

```python
get_courses_Udacity = soup_Udacity.find_all('a', {'class':'capitalize'})
```

```python
courses_Udacity = []
for course in get_courses_Udacity:
    courses_Udacity.append(course.text)
courses_Udacity
```

```python
len(courses_Udacity)
```

```python
data_courses_Udacity = pd.DataFrame(courses_Udacity)
data_courses_Udacity.head(245)
```

```python
data_courses_Udacity.to_csv('courses_Udacity.csv')
```

```python
get_programs_Udacity = soup_Udacity.find_all('h4', {'class':'category ng-star-inserted'})
programs_Udacity = []
for course in get_programs_Udacity:
    programs_Udacity.append(course.text)
programs_Udacity = set(programs_Udacity)
programs_Udacity
```

```python
len(programs_Udacity)
```

```python
data_programs_Udacity = pd.DataFrame(programs_Udacity)
data_programs_Udacity.head(7)
data_courses_Udacity.to_csv('programs_Udacity.csv')
```
