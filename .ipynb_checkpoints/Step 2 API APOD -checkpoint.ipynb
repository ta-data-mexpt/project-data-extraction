```python
import json
import requests
import pandas as pd
import  datetime 
import  urllib.request
```

```python
#Fechas a examinar API = fecha especifica en url

#Parametros 
fecha_hoy='2020-05-13'
longitud_web_scrapping=1960
key='L7KGP9ToI5xwQRiwiYjoj8z1RvqWMhfgAr3QQaWQ'

dates= pd.date_range(end =fecha_hoy, periods=longitud_web_scrapping).to_pydatetime().tolist()


#Convertir formato de fecha a string

list_dates=[] 
for date in dates:
#    print(date)
    year = date.strftime("%Y")
#    print("year:", year)
    month = date.strftime("%m")
#   print("month:", month)
    day = date.strftime("%d")
#  print("day:", day)
    date_string= date.strftime("%Y-%m-%d")
    list_dates.append(date_string)

#Modificando las urls con la fecha urls de todas las apod 2019 + la key_api

urls_api= []
for date in list_dates:
    apod='https://api.nasa.gov/planetary/apod?date='+date+'&hd=True&api_key='+key
    urls_api.append(apod)

#Verificar que ultimo valor de fecha sea día actual debe estar en sincronia con Web_Scrapping
urls_api[-1]
```

```python
#Corroborar longitud de lista sea igual a longitud_web_scrapping

len(urls_api)
```

```python
# Extrae json de cada link con la API 

APOD=[]
for information in urls_api:
    request_info= requests.get(information)
    resultado=request_info.json()
    APOD.append(resultado)
```

```python
#Data frame     
df1_request=pd.DataFrame(APOD)
df1_request
```

```python
#Data frame to csv 

df1_request.to_csv('Api_nasa_apod_information.csv')
```
