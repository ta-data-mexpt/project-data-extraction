```python
def pags_a_revisar_ML():
    #preguntamos las paginas a scrapear
    print("¿Cuántas páginas de las ofertas de Mercado Libre quieres scrapear del 1 al 10?")
    flag=True
    while flag:
        try:
            #Recibimos la respuesta en una variable y la pasamos a entero
            pags = int(input())
            if pags in range(1,11):
                flag=False
                
            else:
                #En caso que no esté dentro del rango permitido
                raise ValueError 
        except ValueError:
                print("Debe ser un número dentro del rango del 1 al 10")           
    return pags 
```

```python
def links_de_productos(page_source):
    #Hacemos la sopa para poder ver mejor la info
    ofertas_mercadolib = BeautifulSoup(page_source,'lxml')
    #En este caso viene como "li class" es por eso que viene el li. y el nombre del item
    link_to_product = ofertas_mercadolib.select('li.promotion-item a[href]')
    #Guardamos los links de los productos para scrapear
    links = [str(link['href']) for link in link_to_product]
    #Guardamos el numero de articulos por pagina que vienen
    articulos_x_pag=len(links)
    return (links,articulos_x_pag)
    
```
