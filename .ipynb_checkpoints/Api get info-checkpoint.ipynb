```python
# obtenemos los titulos del campeón
titulos = []
for i in campeones:
    #print(champs['data'][i]['title'])    
    titulos.append(champs['data'][i]['title'])
```

```python
camp_titulos = []
for i,j in enumerate(campeones):
    # j + '\n' + titulos[i]
    # print(camp_titulos)
    camp_titulos.append(j + '-' + titulos[i] + '.')
print(camp_titulos[0])
print(len(camp_titulos))
```

```python
# obtenemos los basic stats de los campeones
# basic_stats = {}
basic_stats = []
for i in campeones:
    #print(champs['data'][i]['info'])
    basic_stats.append(champs['data'][i]['info'])
basic_stats[0]
```

```python
# obtenemos un diccionario de campeones con sus stats
champ_title = {}
for i,j in enumerate(campeones):
    #print(i)
    #print(j)
    champ_title[j] = titulos[i]
```

```python
#champ_title
```

```python
# obtenemos un diccionario de campeones con sus stats
champ_info = {}
for i,j in enumerate(campeones):
    #print(i)
    #print(j)
    champ_info[j] = champs['data'][j]['info']
```

```python
# Aquí obtenemos la lista del campeón con ESTADÍSTICAS BÁSICAS
info_champ = list(zip(camp_titulos,basic_stats))
#list(info_champ)
info_champ[0]
```

```python
# obtenemos el ROL de cada campeón
tags = []
for i in campeones:
    #print(champs['data'][i]['tags'])
    tags.append(champs['data'][i]['tags'])
    #print(type(champs['data'][i]['tags']))
```

```python
# añadiremos el rol principal de cada campeón
tabl_champs = list(zip(info_champ,tags))
tabl_champs[0]
#print(len(tabl_champs))
```

```python
# buscaremos la popularidad de un campeón mediante el número de 'skins' que tiene
# el campo de skins no se encuentra en la URL anteriror ocupada
# se debe entrar en 'http://ddragon.leagueoflegends.com/cdn/10.9.1/data/es_MX/champion/<champion_name>.json'
url_champ = []
for i in campeones:
    #print('http://ddragon.leagueoflegends.com/cdn/10.9.1/data/es_MX/champion/%s.json'%(i))
    url_champ.append('http://ddragon.leagueoflegends.com/cdn/10.9.1/data/es_MX/champion/%s.json'%(i))
url_champ[1]
```

```python
# accedemos a la base de datos de cada campeón y guardamos el json
chracters_json = []
for i in url_champ:
    personaje_get = requests.get(i)
    campion_info = personaje_get.json()
    # print(champs['data'][i]['skins'])
    chracters_json.append(campion_info)
```

```python
skins_per_character = []
count = 0
for i in chracters_json:
    for j in campeones:
        #print(i)
        #print(j)
        try:
            print(len(i['data'][j]['skins']))
            print(i['data'][j]['skins'])
            count += 1
        except:
            pass
        break
```

```python
# Faltara implementar los nombres de cada skin
skins_per_character = []
#skins_name = []
#count = 0
for i in chracters_json:
    for j in campeones:
        #print(i)
        #print(j)
        try:
            #print(len(i['data'][j]['skins']))
            skins_per_character.append(len(i['data'][j]['skins']))
            #print(i['data'][j]['skins'][count]['name'])
            #count += 1
        except:
            pass
```
