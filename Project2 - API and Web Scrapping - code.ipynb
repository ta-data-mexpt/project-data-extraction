```python
#Step 2: Import and Set Up (incl. Spotipy) 
import spotipy
import os
import sys
import webbrowser
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials #To access authorised Spotify data
import spotipy.oauth2 as oauth2
import spotipy.util as util
import json
from json.decoder import JSONDecodeError
import requests
import numpy as np
```

```python
# Step 3: Authorize API calls:
with open('client.txt') as f:
    client_id = f.readlines()
    client_id = client_id[0].split()
    client_id = client_id[0]

with open('secret.txt') as s:
    secret_id = s.readlines()
    secret_id = secret_id[0]
```

```python
#Step 4: Create a Spotify object to call methods:
token = oauth2.SpotifyClientCredentials(
    client_id=client_id, 
    client_secret=secret_id)
cache_token = token.get_access_token()
sp = spotipy.Spotify(cache_token)
```

```python
#Step 5: Retrieve Data from Spotify Web API: Tracks General Data
artist_name = []
track_name = []
popularity = []
track_id = []
#genres = [] 
for i in range(0,10000,50):
    track_results = sp.search(q='year:2020', type='track', limit=50,offset=i, market= 'MX')
    for i, t in enumerate(track_results['tracks']['items']):
        artist_name.append(t['artists'][0]['name'])
        track_name.append(t['name'])
        track_id.append(t['id'])
        popularity.append(t['popularity'])
        #genres.append(t['genres'])
```

```python
#Step 6: Load Data into DataFrame
import pandas as pd
```

```python
track_dataframe = pd.DataFrame({'artist_name' : artist_name, 'track_name' : track_name, 'track_id' : track_id, 'popularity' : popularity})
print(track_dataframe.shape)
track_dataframe.head()
```

```python
#Step 7: Sort by popularity and genre in Mexico
track_dataframe_sorted=track_dataframe.sort_values(by='popularity', ascending = False)
print(track_dataframe_sorted.shape)
track_dataframe_sorted.head()
```
