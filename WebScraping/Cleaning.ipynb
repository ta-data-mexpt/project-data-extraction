```python
#Añadiendo en un solo data frame los datos brutos obtenidos del scrapping.
datos =pd.DataFrame()
for i in range(9):
    nombre_archivo = 'Datos_SIEM{}.csv'.format(i)
    datos_nuevos = pd.read_csv(nombre_archivo, usecols=range(0,4),na_values = ['-'])
    datos = pd.concat([datos, datos_nuevos])
datos.reset_index(drop = True,inplace =True)
datos.columns = columns_datos
datos
```

```python
#Generando una columna con ID único 
id = [i for i in range(1,91)]
datos['ID']= id
```

```python
#Trabajando con la primera lista de información Datos Generales
#Limpiando la lista 
print('longitud',len(datos['Datos Generales'][0]))
print( 'tipo de datos',type(datos['Datos Generales'][0]))
lista = datos['Datos Generales'][0].split(',')
#Sustituyendo las comillas qeu tiene de más
lista_limpia = [re.sub(r"\'"," ", element) for element in lista]
#Quitando espacios en blanco 
lista_limpia = [element.strip() for element in lista_limpia]
print(lista_limpia)
print(len(lista_limpia))
```

```python
#Buscando las posiciones en lista que tiene información relevante
#Lista con la información que se quiere extraer de los datos brutos de Datos Generales
datos_generales= ['Nombre Comercial','Tipo de vialidad','Nombre de la vialidad','Número Exterior',\
        'Edificio. Piso o Nivel','Número Interior','Número de Local','Código Postal',\
        'ASENTAMIENTO HUMANO','ASENTAMIENTO HUMANO','PARQUE INDUSTRIAL U OTRO',\
        'PARQUE INDUSTRIAL U OTRO', 'Tipo 1','Nombre 1','Tipo 2','Nombre 2','VIALIDAD POSTERIOR',\
        'VIALIDAD POSTERIOR','Descripción de ubicación','Entidad Federativa','Delegación o Municipio',\
        'Localidad', 'Teléfono', 'Extensión','E-mail','Otro','Sitio Web']
#Generando una lista con la posicion -1 de los datos de interes
#hay pocos casos donde se pone la posción -2 o -4, cuando el indicador ser repite.
indices_datos_generales = []
for element in datos_generales :
    indices_datos_generales.append(lista_limpia.index(element))
#len(indices_datos_generales)
```

```python
#Sumando la posición que le correponde para obtener la posición de interés.
posiciones_interes = [0] * (len(indices_datos_generales))
posiciones_interes[0] = indices_datos_generales[0] + 1
for i in range(1,len(indices_datos_generales)):
    #print('{}\n',i)
    if i == len(indices_datos_generales)-1:
        posiciones_interes[i] = indices_datos_generales[i]+1
    elif indices_datos_generales[i-1] == indices_datos_generales[i]:
        posiciones_interes[i]=indices_datos_generales[i]+4
    elif indices_datos_generales[i+1] != indices_datos_generales[i]:
        posiciones_interes[i]=indices_datos_generales[i]+1
    else :  
        posiciones_interes[i]=indices_datos_generales[i]+2
#posiciones_interes
```

```python
#Generando una lista con los datos de de interes.
datos_generales_columns = [datos_generales[element-1].strip() for element in posiciones_interes]
datos_generales = pd.DataFrame(columns=datos_generales_columns)
```
