### Top 100 rated anime - MyAnimeList

* Esto es una lista de los top 100 anime en MyAnimeList.net

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
top50 = 'https://myanimelist.net/topanime.php'
```

```python
top50req = requests.get(top50).content
```

```python
top50soup = BeautifulSoup(top50req, 'lxml')
top50soup
```

```python
top50anime_name = top50soup.select('h3[class="hoverinfo_trigger fl-l fs14 fw-b anime_ranking_h3"]')
top50anime_name
```

```python
top_50_names = [name.text for name in top50anime_name]
top_50_names
```

```python
top50anime_score = top50soup.select('div[class="js-top-ranking-score-col di-ib al"]')
top50anime_score
```

```python
top_50_score = [score.text for score in top50anime_score]
top_50_score
```

```python
top100 = 'https://myanimelist.net/topanime.php?limit=50'
```

```python
top100req = requests.get(top100).content
```

```python
top100soup = BeautifulSoup(top100req, 'lxml')
top100soup
```

```python
top100anime_name = top100soup.select('h3[class="hoverinfo_trigger fl-l fs14 fw-b anime_ranking_h3"]')
top100anime_name
```

```python
top_100_names = [name.text for name in top100anime_name]
top_100_names
```

```python
top100anime_score = top100soup.select('div[class="js-top-ranking-score-col di-ib al"]')
top100anime_score
```

```python
top_100_score = [score.text for score in top100anime_score]
top_100_score
```

```python
top_100_anime = top_50_names + top_100_names
top_100_anime
```

```python
top_100_scores = top_50_score + top_100_score
top_100_scores
```

```python
anime_df = pd.DataFrame({'Anime': top_100_anime, 'Score': top_100_scores})
anime_df
```

```python
anime_df.to_csv('Top_100_anime.csv')
```
